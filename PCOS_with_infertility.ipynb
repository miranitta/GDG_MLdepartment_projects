```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Chargement des donn√©es
di = pd.read_csv("PCOS_infertility.csv")

# Suppression des espaces multiples dans les noms des colonnes
di.columns = di.columns.str.replace(r'\s+', ' ', regex=True).str.strip()

# V√©rification des noms des colonnes apr√®s nettoyage
print("Colonnes disponibles apr√®s nettoyage :", di.columns.tolist())

# Copie du DataFrame pour √©viter de modifier l'original
di1 = di.copy()

# V√©rification des valeurs uniques dans la colonne 'AMH(ng/mL)'
print("Valeurs uniques dans 'AMH(ng/mL)':", di1['AMH(ng/mL)'].unique())

# Remplacement des valeurs non num√©riques par NaN
di1['AMH(ng/mL)'] = pd.to_numeric(di1['AMH(ng/mL)'], errors='coerce')

# Suppression des lignes contenant des NaN dans 'AMH(ng/mL)'
di1 = di1.dropna(subset=['AMH(ng/mL)'])

# V√©rification apr√®s conversion
print("Donn√©es apr√®s nettoyage :", di1.info())

# Visualisation : Distribution des niveaux de I beta-HCG
plt.figure(figsize=(8, 6))
sns.histplot(data=di1, x='I beta-HCG(mIU/mL)', kde=True, color='skyblue')
plt.title('Distribution de I beta-HCG (mIU/mL)')
plt.xlabel('I beta-HCG (mIU/mL)')
plt.show()

# Visualisation de la relation entre I et II beta-HCG
plt.figure(figsize=(8, 6))
sns.scatterplot(x='I beta-HCG(mIU/mL)', y='II beta-HCG(mIU/mL)',
                data=di1, hue='PCOS (Y/N)', palette='Set1')
plt.title('Relation entre I beta-HCG et II beta-HCG')
plt.xlabel('I beta-HCG (mIU/mL)')
plt.ylabel('II beta-HCG (mIU/mL)')
plt.legend(title='PCOS')
plt.show()

# Visualisation avec boxplot pour AMH(ng/mL) en fonction du statut PCOS
plt.figure(figsize=(8, 6))
sns.boxplot(x='PCOS (Y/N)', y='AMH(ng/mL)', data=di1, palette='Set2')
plt.title('AMH (ng/mL) par statut PCOS')
plt.xlabel('PCOS (0 = Non, 1 = Oui)')
plt.ylabel('AMH (ng/mL)')
plt.show()
```
